```python
import io

from apiclient.http import MediaIoBaseDownload


def get_client(keyfile=None):
    scopes = ['https://www.googleapis.com/auth/devstorage.read_only']
    if keyfile:
        scoped_credentials = ServiceAccountCredentials.from_json_keyfile_name(keyfile, scopes)
    else:
        credentials = GoogleCredentials.get_application_default()
        scoped_credentials = credentials.create_scoped(scopes)
    http = httplib2.Http()
    scoped_credentials.authorize(http)
    return discovery.build('storage', 'v1', http=http)


def bucket_prefix(full_path):
    path = full_path.replace('gs://', '').split('/')
    bucket_name = path[0]
    prefix = '/'.join(path[1:])
    return bucket_name, prefix

def get_storage_size(prefix, client=None):
    if not client:
        client = get_client()
    bucket_name, prefix = bucket_prefix(prefix)
    resp = client.objects().list(bucket=bucket_name, prefix=prefix).execute()
    if 'items' in resp:
        return sum(int(x['size']) for x in resp['items'])
    else:
        return 0

def list_files(prefix, client=None):
    if not client:
        client = get_client()
    bucket_name, prefix = bucket_prefix(prefix)
    resp = client.objects().list(bucket=bucket_name, prefix=prefix).execute()
    if 'items' in resp:
        return ['{0}/{1}'.format(x['bucket'], x['name']) for x in resp['items']]
    else:
        return []
```
